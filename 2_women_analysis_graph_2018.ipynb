# Study of women participation in higher education in 2018

```python
import pandas as pd
import numpy as np
from matplotlib.pyplot import *

df_total = pd.read_csv("data/2018.csv",sep="|", dtype='unicode')

df = df_total.query('TP_SEXO == "1"')

df[:10]
```
